### Single Variable Derivative

**General Definition:**

for a function $y = f(x)$, its derivative is defined as
${\displaystyle f'(x)=\lim _{h\to 0}{\frac {f(x+h)-f(x)}{h}}}$. A function is differentiable at $x_0$ if such a limit exits at $x=x_0$.

the second order derivative, $f''(x)$, of a function $f(x_0)$ is the derivative of the derivative of $f(x)$

$f''(x)=\lim _{h\to 0}{\frac {f(x+h)-2f(x)+f(x-h)}{h^{2}}}$



**Geometric Definition:**

The derivative $f'
(x_0)$ of $f$ at $x_0$ is the slope of the tangent line ("just touch")
to $y = f(x)$ at the point $P = (x_0, f(x_0))$

<img src="0_Tangent_to_a_curve.svg.png" width="400">

**Meanings and Derivative tests**
- First derivative

    The first derivative can be interpreted as an instantaneous rate of change. So it tells us about the direction the function is going, that if the function is increasing or decreasing.

    - If $f'(x)>0$, $f$ is increasing at $x$
    - If $f'(x)<0$, $f$ is decreasing at $x$
    - If $f'(x)=0$, $f$ has a local(can be global) maximum or minimum value (relative extremum point), or an inflection point at $x$

- Second derivative

     The second derivative measures how the rate of the change of the function changes.
     
     For $f(x)$, if on $x\in (a, b)$:
     - $f''(x)<0$, then $f$ is concave donw on $x\in (a, b)$ (the line formed by any two points on the function is below the curve) 
     - $f''(x)>0$, the $f$ is concave up on $x\in (a, b)$ (the line formed by any two points on the function is above the curve). 
     
     >
     
     For a point where $f'(x)=0$:
     - If $f''(x)<0$, then $f$ has a local maximum at $x$
     - If $f''(x)>0$, then $f$ has a local minimum at $x$
     - If $f''(x)=0$, the second derivative test says nothing about the point $x$, a possible inflection point, where the function changes concavity.
     

- Example 1:

    For $f(x) = x^2$, it's first derivative is $f'(x)=2x$, and second derivative $f''(x)=2$
    At x = 0, $f'(0) = 0$ and $f''(0)=2$, so $f(x)$ has a local minimum at $x=0$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.arange(-3.0, 3.0, 0.01)
y = x**2

plt.plot (x, y,label = '$f(x)$')
plt.hlines(y=0, xmin=-3, xmax=3, colors='g',label='tangent line at x= 0')
plt.plot([0],[0],'ro',label='f(x)=0')
plt.legend()

plt.show()
```

- Example 2:

    For $f(x) = x^3$, it's first derivative is $f'(x)=3x^2$, and second derivative $f''(x)=6x$
    At x = 0, $f'(0) = 0$ and $f''(0)=0$. From the plot we can see $f(0)=0$ is an inflection point

**Exercise**

Find the relative extremum points for function
$f(x)=\frac{x^2}{x-1}$

To find the relative extremum points, we first need to find at where the first derivative is 0
We use the quotient rule here to find $f'(x)$

$f'(x) = \frac{(x-1)*(x^2)'-(x-1)'*(x^2)}{(x-1)^2}$

$=\frac{(x-1)2x-(x^2)}{(x-1)^2}$

$=\frac{2x^2-2x-x^2}{(x-1)^2}$

$=\frac{x^2-2x}{(x-1)^2}$

$=\frac{x(x-2)}{(x-1)^2}$

So the extremum points are at $x=0$ and $x=2$

Also note that this function is not defined at x=1

### Partial Derivative

For a multivariable function, a partial derivative is its derivative with respect to one of those variables, with the others held constant.

For example, for $z=f(x,y)$, the first order partial derivative of
$z$ with respect to x is:

$f_x(x, y) = z_x = {\frac{\partial f}{\partial x}}$

Similarly, for partial with respect to $y$:

$f_y(x, y) = z_y = {\frac{\partial f}{\partial y}}$

The second order derivatives:

$f_{xx}(x, y) = z_{xx} = \frac{\partial^2 f}{\partial x^2}$

$f_{yy}(x, y) = z_{yy} = \frac{\partial^2 f}{\partial y^2}$

The cross partial derivatives:

$f_{xy}(x, y) = f_{yx}(x, y) =z_{xy} = \frac{\partial^2 f}{\partial x \partial y}$

**Gradient**

The vector $<f_{x},f_{y}>$ is called the gradient $\nabla f$

**Example**

Find the first, second, and cross partiall derivatives for $f(x,y) = x^2+5xy+2y^2$

- First order

    $f_x(x, y) = {\frac{\partial f}{\partial x}}= 2x+5y$
    
    $f_y(x, y) = {\frac{\partial f}{\partial y}} = 5x+4y$
    
- Second order

    $f_{xx}(x, y) = \frac{\partial^2 f}{\partial x^2} = 2$
    
    $f_{yy}(x, y) = \frac{\partial^2 f}{\partial y^2} = 4$
        
- Cross

    $f_{xy}(x, y) = \frac{\partial}{ \partial y}(\frac{\partial f}{ \partial x}) = 5$
    
    $f_{yx}(x, y) = \frac{\partial}{ \partial x}(\frac{\partial f}{ \partial y}) = 5$

**Relative extrema and saddle points**
- $f(x,y)$ is at a relative extremum or saddle points if $\nabla f =<0,0>$

A saddle point is at where the function has a local maximum in one direction, but a local minimum in another direction
<img src="3_Saddle_point.svg.png" width="400">

- Second derivative test

    At the point $(x_0,y_0)$ where $\nabla f(x_0,y_0) =<0,0>$\
    We define the Hessian $H$ of $f(x,y)$ as $f_{xx}(x_0,y_0)f_{yy}(x_0,y_0)- {f_{xy}(x_0,y_0)}^2$
    
    - if $H<0$, it's a saddle point
    - if $H>0$ and $f_{xx}<0$, it's a local maximum
    - if $H>0$ and $f_{xx}>0$, it's a local minimum
    - if $H=0$ we can't say anything about the point
    
- Exercise
    
    Find the relative extrema points for $f(x,y)=x^4+y^4-4xy+1$

To find the relative extrema points, we need to find at where $\nabla f(x_0,y_0) =<0,0>$ 

So we take the first order partial derivative respects to $x$ and $y$, and set them to 0

$f_x = 4x^3 - 4y$

$f_y = 4y^3 - 4x$

That gives us

$y=x^3$

$x=y^3=(x^3)^3=x^9$

Fractorize and reorganize this expression we got

$0 = x(x+1)(x-1)(x^2+1)(x^4+1)$

Thus $x = 0, 1, -1$

So the relative extrema points are at $(0,0)$, $(1,1)$ and $(-1,-1)$


### Taylor series
Taylor series is an infinite sum of terms to approximate any function as a polynomial, that are expressed in terms of the function's derivatives at a single point.

**Single variable**

the Taylor series for a single variable function is ${\displaystyle f(a)+{\frac {f'(a)}{1!}}(x-a)+{\frac {f''(a)}{2!}}(x-a)^{2}+{\frac {f'''(a)}{3!}}(x-a)^{3}+\cdots ,}$

More formally, we write it as ${\displaystyle \sum _{n=0}^{\infty }{\frac {f^{(n)}(a)}{n!}}(x-a)^{n},}$
where $f(n)(a)$ is the nth derivative of $f$ evaluated at the point $a$

When a = 0, the series is also called a Maclaurin series

In practice, the series to the second order term usually already gives us a good approximation

**Example: the Maclaurin series for the exponential function $e^x$**

$f^{(n)}{(a)}$ for $e^x$ is $e^x$, $f^{(n)}(a) = 1$ when $a = 0$

Thus the Maclaurin series is simply

${\displaystyle e^{x}=\sum _{n=0}^{\infty }{\frac {x^{n}}{n!}}=1+x+{\frac {x^{2}}{2!}}+{\frac {x^{3}}{3!}}+\cdots }$

**Exercise:**

find the Maclaurin series for $\frac 1 {1-x}$ to the second order

For Maclaurine series, $a =0$

The $0th$ order term:

$f(a)=\frac 1 {1-x}=\frac 1 {1-0}=1$

The $1st$ order term:

${\frac {f'(a)}{1!}}(x-a) = \frac 1 {(1-a)*1!}x = x$

The $2nd$ order term:

${\frac {f''(a)}{2!}}(x-a)^2 = \frac 2 {(1-a)*2!}x^2 = x$

So the Taylor series to the second order is $1+x+x^2$

The full infinite series is ${\displaystyle \sum _{n=0}^{\infty }x^{n},}$
the geometric series

**Two variables**

$f(x,y)=\sum_{n=0}^{\infty} \frac 1 {n!}\left(x\frac {\partial}{\partial x}+y\frac {\partial}{\partial y} \right)^n f(x,y)|_{(x,y)=(x_0,y_0)}$
$ =f(x_0,y_0)+(x-x_0)f_x(x_0,y_0)+(y-y_0)f_y(x_0,y_0)+\frac 1 2(x-x_0)^2f_{xx}(x_0,y_0)+(x-x_0)(y-y_0)f_{xy}(x_0,y_0)+\frac 1 2 (y-y_0)^2f_{yy}(x_0,y_0) \dots$



### Lagrange Multiplier

- The Lagrange multiplier is a technique to find the maximum or minimum of a multivariable function $f(x,y,\cdots)$ when there is some constraint on the input values
- The constraint has to be in the form $g(x,y,\cdots) = c$ where $g(x,y,\cdots)$ is a multivariable function of the same input space as $f$, and $c$ a constant
- In order to find the maximum or minimum of a function ${\displaystyle f(x,y,\cdots)}$ subjected to the equality constraint ${\displaystyle g(x,y,\cdots)=c}$, form the Lagrangian function

${\displaystyle {\mathcal {L}}(x,y,\cdots,\lambda )=f(x,y,\cdots)+\lambda (g(x,y,\cdots)-c)}$\
and find the extremum value for the Lagrangian function

**Example**

Find the extremum points for ${\displaystyle f(x,y)=x^{2}y} $
under the constrain ${\displaystyle x^{2}+y^{2}=1}$


Let ${\displaystyle g(x,y)=x^{2}+y^{2}-1}$

and set the Lagrangian function

$ {\displaystyle \mathcal {L}(x,y,\lambda )=f(x,y)+\lambda g(x,y)=x^{2}y+\lambda (x^{2}+y^{2}-1)}$

We set the all the first order partial derivatives to zero and the solutions are the extremum points


$\mathcal {L}_x = 2 x y + 2 \lambda x = 0$

$\mathcal {L}_y = x^2 + 2 \lambda y = 0$

$\mathcal {L}_\lambda = x^2 + y^2 -1 = 0$



**Exercise**

Find the maximum and minimum of $f\left( {x,y} \right) = 5x - 3y$ subject to the constraint 
${x^2} + {y^2} = 136$

Let $g(x,y)= x^2+y^2-136$

and set the Lagrangian function

$ {\displaystyle \mathcal {L}(x,y,\lambda )=f(x,y)+\lambda g(x,y)=5x-3y+\lambda (x^{2}+y^{2}-136)}$

Then set the all the first order partial derivatives to zero and the solutions are the extremum points


$\mathcal {L}_x = 5 + 2 \lambda x = 0$

$\mathcal {L}_y = -3 + 2 \lambda y = 0$

$\mathcal {L}_\lambda = x^2 + y^2 -136 = 0$


We get

$x = -\frac{5}{{2\lambda }}$

$y =  \frac{3}{{2\lambda }}$

Then

$\frac{{25}}{{4{\lambda ^2}}} + \frac{9}{{4{\lambda ^2}}} = \frac{{17}}{{2{\lambda ^2}}} = 136$

Solve for $\lambda$

${\lambda ^2} = \frac{1}{{16}}\hspace{0.25in} \Rightarrow \hspace{0.25in}\,\,\,\,\,\lambda  =  \pm \frac{1}{4}$

Plug $\lambda$ back to $x$ and $y$

for $\lambda  =  - \frac{1}{4}$

$x=-10$

$y=6$

for $\lambda  =   \frac{1}{4}$

$x=10$

$y=-6$

So extrema points are at $(10,-6)$ and $(-10,6)$ (no saddle point here)
